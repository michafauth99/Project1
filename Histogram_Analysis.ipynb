```python
import cv2
import matplotlib.pyplot as plt
import sys
import os
import numpy as np
import plotly.graph_objects as go


# Add the 'src' directory to the Python path
sys.path.append(os.path.join(os.getcwd(), 'src'))
```

```python
# define sand patch coordinates to measure average intensity
x=600
y=900
size=100
```

```python
def plot_histograms():
    fig = go.Figure()  # Create a figure for the combined plot
    
    for filename in os.listdir('data/images'):
        img = cv2.imread(os.path.join('data/images', filename), cv2.IMREAD_GRAYSCALE)
        hist = cv2.calcHist([img], [0], None, [256], [0, 256]).flatten()  # Flatten to 1D array for Plotly

        # calculate the average pixel value for the whole image
        avg_img = np.mean(img)
        print(f"Average pixel value for {filename}: {avg_img}")
        
        # calculate the average pixel value for the patch
        patch = img[y:y+size, x:x+size]
        avg_patch = np.mean(patch)
        print(f"Average pixel value for the sand patch in {filename}: {avg_patch}")
        
        # Add the histogram as a trace to the figure
        fig.add_trace(go.Scatter(x=np.arange(256), y=hist, mode='lines', name=filename))

    # Update the layout to add title and axis labels
    fig.update_layout(
        title="Histograms of All Images",
        xaxis_title="Pixel Intensity",
        yaxis_title="Frequency",
        showlegend=True
    )
    
    fig.show()  # Show the interactive plot

plot_histograms()
```
